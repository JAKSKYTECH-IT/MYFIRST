## Once I have the GameStop stock data, I will use the make_graph function to plot it along with providing a title for the graph. Here's how i will do it:

```python
import matplotlib.pyplot as plt
```

```python
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='Close Price', color='red')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```

```python
# Call the make_graph function with the GameStop stock data and a title
```

```python
make_graph(gme_data, title="GameStop Stock Price (2021)")
```
