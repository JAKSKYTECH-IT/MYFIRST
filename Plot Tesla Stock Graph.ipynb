### Once I have the Tesla stock data, I will use the make_graph function to plot it along with providing a title for the graph. Here's how I will do it:

```python
import matplotlib.pyplot as plt
```

```python
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='Close Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```

```python
# Call the make_graph function with the Tesla stock data and a title
```

```python
make_graph(tesla_data, title="Tesla Stock Price (2021)")
```
